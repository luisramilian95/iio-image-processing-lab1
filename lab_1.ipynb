```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
R, G, B = (0, 1, 2)
```

```python
def cv2_imshow(image) :
    plt.imshow(image)
    plt.show()
```

```python
def change_image_color(image, channel):
  
  new_image = image.copy()

  # Blue
  # set green and red channels to 0
  if channel == 1:
    new_image [:, :, G] = 0
    new_image [:, :, R] = 0

  # Green
  # set blue and red channels to 0
  if channel == 2 :
    new_image[:, :, B] = 0
    new_image[:, :, R] = 0

  # Red
  # set blue and green channels to 0
  if channel == 3 :
    new_image[:, :, B] = 0
    new_image[:, :, G] = 0

  # Red and Green
  # set blue channel to 0
  if channel == 10:
    new_image[:, :, B] = 0

  # Green and Blue
  # set red channel to 0
  if channel == 20:
    new_image[:, :, R] = 0

  # Blue and Red
  # set green channel to 0
  if channel == 30:
    new_image[:, :, G] = 0

  return new_image
```

```python
image = cv2.imread('images/gfg.jpeg')
cv2_imshow(change_image_color(image, 30))
```

```python
def build_image(image_blue, image_red, image_green):
  new_image = image_blue.copy()
  height = len(new_image)
  width = len(new_image[0])

  for pixel_position_y in range(height):
    for pixel_position_x in range(width):

      blue = image_blue[pixel_position_y][pixel_position_x][B]
      green = image_green[pixel_position_y][pixel_position_x][G]
      red = image_red[pixel_position_y][pixel_position_x][R]

      new_image[pixel_position_y][pixel_position_x] = [red, green, blue]

  return new_image
```

```python
image_gray_blue = cv2.imread('images/imagen1/imagen1_salida_gray_azul.jpg')
image_gray_red = cv2.imread('images/imagen1/imagen1_salida_gray_rojo.jpg')
image_gray_green = cv2.imread('images/imagen1/imagen1_salida_gray_verde.jpg')

new_image = build_image(image_gray_blue, image_gray_red, image_gray_green)
cv2_imshow(new_image)
```

```python
image_gray_blue = cv2.imread('images/imagen3/perro_salida_gray_azul.jpg')
image_gray_red = cv2.imread('images/imagen3/perro_salida_gray_rojo.jpg')
image_gray_green = cv2.imread('images/imagen3/perro_salida_gray_verde.jpg')

new_image = build_image(image_gray_blue, image_gray_red, image_gray_green)
cv2_imshow(new_image)
```

```python
def build_3D_gray_scale_image(image) :

  image_blue = image.copy()
  image_green = image.copy()
  image_red = image.copy()

  height = len(image)
  width = len(image[0])

  for pixel_position_y in range(height):
    for pixel_position_x in range(width):

      pixel = image[pixel_position_y][pixel_position_x]

      red = pixel[2]
      green = pixel[1]
      blue = pixel[0]

      image_blue[pixel_position_y][pixel_position_x] = [blue, blue, blue]
      image_green[pixel_position_y][pixel_position_x] = [green, green, green]
      image_red[pixel_position_y][pixel_position_x] = [red, red, red]


  return image_blue, image_green, image_red
```

```python
image_blue, image_green, image_red = build_3D_gray_scale_image(image)
```

```python
new_image = build_image(image_blue, image_red, image_green)
cv2_imshow(new_image)
```

```python
def calc_histogram(image):
  new_image = image.copy()
  height = len(new_image)
  width = len(new_image[0])

  blue_data = []
  green_data = []
  red_data = []
  gray_scale = []

  for pixel_position_y in range(height):
    for pixel_position_x in range(width):

      pixel = new_image[pixel_position_y][pixel_position_x]

      blue_data.append(pixel[B])
      green_data.append(pixel[G])
      red_data.append(pixel[R])
      gray_scale.append(np.average(pixel))


  plt.title('Histograma Azul')
  blue_histogram, bins = np.histogram(blue_data, bins=range(256))
  plt.plot(blue_histogram, 'blue')
  plt.show()

  plt.title('Histograma Verde')
  green_histogram, bins = np.histogram(green_data, bins=range(256))
  plt.plot(green_histogram, 'green')
  plt.show()


  plt.title('Histograma Rojo')
  red_histogram, bins = np.histogram(red_data, bins=range(256))
  plt.plot(red_histogram, 'red')
  plt.show()

  plt.title('Histograma Escala de Grises')
  gray_histogram, bins = np.histogram(gray_scale, bins=range(256))
  plt.plot(gray_histogram, 'black')
  plt.show()
```

```python
calc_histogram(image)
```
